```python
### Label  文字
### Button 按鈕
### 用 Button 控制顯示

import tkinter as tk

window = tk.Tk()
window.title("my window")
window.geometry("200x100")

var = tk.StringVar()
#Label = tk.Label(window, text="OMG! this is TK!", bg="green", font=("Arial",12), width=15, height=2)
Label = tk.Label(window, textvariable=var, bg="green", font=("Arial",12), width=15, height=2)
Label.pack()

on_hit = False
def hit_me():
    global on_hit
    if on_hit == False:
        on_hit = True
        var.set("you hit me")
    else:
        on_hit = False
        var.set("")
    
Button = tk.Button(window, text="hit me", width=15, height=2, command=hit_me)
Button.pack()
window.mainloop()
```
